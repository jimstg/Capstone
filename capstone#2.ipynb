```python
#Here I am looking at the size of the group of employees that are lost due to attrition and those that are not.
print("Employee Attrition, Yes: ", len(df[df.Attrition == 'Yes']))
print("Employee Attrition No: ", len(df.Attrition == 'No'))

#This is the mean of the Job Level for the employees lost to attrition and those that are not.
print('Job Level Mean for Yes: ', df[df.Attrition == 'Yes'].JobLevel.mean())
print('Job Level mean for No: ', df[df.Attrition == 'No'].JobLevel.mean())

#It appears that the higher the job level the less likely you are to leave the company. Let's dig deeper and see if
#this holds true for the total working years as well.
print('Working years for Yes: ', df[df.Attrition == 'Yes'].TotalWorkingYears.mean())
print('Working years for No: ', df[df.Attrition == 'No'].TotalWorkingYears.mean())

#The last two sections of analysis involve variables that really cannot be changed for everyone, so these are not 
#feasible for an A/B style of testing. Now I will compare attrition to environment satisfaction.
print('Environment satisfaction Yes: ', df[df.Attrition == 'Yes'].EnvironmentSatisfaction.mean())
print('Environment satisfaction No: ', df[df.Attrition == 'No'].EnvironmentSatisfaction.mean())

#Checking the relatinship satisfaction with attrition as well.
print('Relationship satisfaction Yes: ', df[df.Attrition == 'Yes'].RelationshipSatisfaction.mean())
print('Relationship satisfaction No: ', df[df.Attrition == 'No'].RelationshipSatisfaction.mean())
```

It appears that environment satisfaction and relationship satisfaction is lower for the employees that leave as a result of attrition. A change in the environment of the company might show a change in attrition.

```python
#The graph shows that those employees that are staying with the company have a higher overall satisfaction with both the 
#environment and the relationships with the company.
fig = df[df.Attrition =='Yes'][['EnvironmentSatisfaction','RelationshipSatisfaction']].groupby('EnvironmentSatisfaction').mean().plot()
df[df.Attrition == 'No'][['EnvironmentSatisfaction','RelationshipSatisfaction']].groupby('EnvironmentSatisfaction').mean().plot(ax=fig)
plt.legend(['Yes','No'])
plt.show()
```

```python
stats.ttest_ind(df[df.Attrition == 'Yes'].EnvironmentSatisfaction,
                df[df.Attrition == 'No'].EnvironmentSatisfaction)
```

Hypothesis: If environmental and relationship satisfaction goes up, then attrition rates will go down.

The pvalue for the relationship between Attrition and Environmental satisfaction is below .05, showing that the "null hypothesis" can not be confirmed. At this point we can dismiss the null hypothesis and continue with the A/B test proposal.

Research Proposal

The Problem - 

Currently there is a higher level of attrition than is acceptable. The value of an employee staying with the company is enormous. Constant turn over and rehiring of new employees costs the company thousands of dollars and reduces consistency in our product. Currently we know that we have a 16% attrition rate. Several factors play into the attrition levels including; experience level, job level, environmental satisfaction, relationship satisfaction. All these variables showed a negative relationship with leaving the company. The longer people stay the less likely they are to leave, the higher the job level the less likely they are to leave. People that stay longer also have a higher average level of satisfaction, both in the environment and in the relationships they have created. This proposal will focus on running a test to see if a change in the environment will have a positive effect on the attrition rate.

The Potential Solution -

To increase environmental satisfaction, we will be making some subtle changes to the office. Fridays will now be "casual" Fridays. Employees will be allowed to wear jeans and non-collared shirts to work. This provides some self-expression in the workplace. We will also place a ping-pong table in the employee breakroom. Once a month there will be an employee "Happy Hour" on a consistent day. This will allow people to interact outside the workday and build meaningful conversations and relationships.

The Method of Testing and Solution -

There will be a general meeting to announce the changes that are being made. After 3 months the staff will be given a survey to check environmental and relationship satisfaction. We will also calculate the attrition rate again at that time. Success would be seeing an attrition rate lower than 10% and higher levels of satisfaction in the workplace.
